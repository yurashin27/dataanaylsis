```python
import matplotlib.pyplot as plt
```

```python
s1=df['SepalLengthCm']
s2=df['SepalWidthCm']
s3=df['PetalWidthCm']
s4=df['PetalLengthCm']
plt.boxplot([s1,s2,s3,s4])

plt.title('Iris Sepal Data')
labels=['sepal_length','sepal_width','petal_width','petal_length']
plt.xticks([1,2,3,4],labels)

plt.show()
```

```python
data=df['SepalLengthCm']
data.plot()
```

```python
plt.hist([s1,s4],label=['sepal_length','petal_length'])
plt.legend()
```

```python
plt.scatter(s1,s2, color='r')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
groups=df.groupby('Species')
```

```python
fig,ax=plt.subplots()
for name, group in groups:
    ax.scatter(group.PetalLengthCm,
            group.PetalWidthCm,
            marker='o',
            label=name)

ax.legend(fontsize=12,loc='upper left')
plt.title('Scatter Plot by Madplotlib', fontsize=15)
plt.xlabel('Petal Length',fontsize=12)
plt.ylabel('Petal Width',fontsize=12)
```
